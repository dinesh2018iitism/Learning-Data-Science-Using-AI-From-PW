np.random.seed(30)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set the random seed for reproducibility
np.random.seed(30)

# Generate the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['X'], df['Y'], df['Z'], c='blue', marker='o')

# Adding labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot')

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales')

sns.set_theme()
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Variation by Month and Day")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the x, y, z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create the figure and a 3D axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surf)

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title('3D Surface Plot')

# Show plot
plt.show()
```

5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

np.random.seed(25)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Setting the seed for reproducibility
np.random.seed(25)

# Creating the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Creating the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'], alpha=0.5)

# Adding titles and labels
plt.title('Bubble Chart of Countries by Population and GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Annotating each bubble with the country name
for i in range(df.shape[0]):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=12, ha='center')

plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set up Bokeh to display plots inline in the notebook
output_notebook()

# Set up data for the sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
from bokeh.models import ColumnDataSource

# Set up Bokeh to display plots inline in the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.random(100) * 10
y = np.random.random(100) * 10
sizes = np.random.randint(10, 50, size=100)
colors = np.random.choice(["red", "blue", "green", "orange", "purple"], size=100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot
p = figure(title="Scatter Plot with Random Data", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Add a scatter renderer with different sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

# Given Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output to a static HTML file
output_file("fruits_bar_chart.html")

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts')

# Add a bar renderer with the data
p.vbar(x=fruits, top=counts, width=0.5 , color= 'red')

# Show the results
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Output to a static HTML file
output_file("histogram.html")

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Data",
           x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer with the data
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Show the results
show(p)
```

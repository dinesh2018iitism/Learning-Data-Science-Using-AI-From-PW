7. What are the advatage of using generators over regular functions

```python
# 1. Memory Efficiency: Generators produce items one at a time and only when required,
# rather than storing the entire sequence in memory.

def generate_numbers(n):
    for i in range(n):
        yield i

# Using a generator to produce a large sequence
gen1 = generate_numbers(1000000)
print (gen1)
```

```python
# 2. Lazy Evaluation:  lazy evaluation, meaning they compute values only when needed
def lazy_squares(n):
    for i in range(n):
        yield i ** 2

# Only computes values as they are iterated over
for square in lazy_squares(10):
    print(square)
```

```python
#3. Simplified Code: Generators can lead to simpler and more readable code.
#4. Infinite Sequences: Generators can represent infinite sequences, which regular functions cannot do efficiently.
```

10. What is the difference between map, reduce and filter functions in Python.

```python
# map(): We need to apply a transformation to each element of an iterable.
# reduce(): We need to aggregate results to a single value, such as summing or multiplying elements.
# filter(): We need to select elements based on a condition.
```

```python
numbers = [1, 2, 3]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

```python
from functools import reduce

numbers = [1, 2, 3]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
numbers = [1, 2, 3, 4, 5]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

1. Write a Python Function that takes a list of numbers as input and returns sum of all even numbers in the list.

```python
def sum_of_evens(numbers):
    total_sum = 0
    for number in numbers:
        if number % 2 == 0:
            total_sum += number
    return total_sum
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(numbers))
```

4. Write a Python function that checks a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

5. Create a Iterator class in python that generate a fibonacci sequence upto a specified number of terms.

```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
for number in fib(10):
    print(number)
```

6. Write a generator functions in Python that yields the powers of 2 upto a given exponent.

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
for power in powers_of_two(5):
    print(power)
```

7. Implement a generator function that reads file line by line and yields each line as a string.

```python
def read_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
file_path = '/content/python.txt'
for line in read_lines(file_path):
    print(line)
```

10. Create a Python program that uses filter to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    def is_not_vowel(char):
        return char not in vowels
    filtered_chars = filter(is_not_vowel, s)
    return ''.join(filtered_chars)
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```
